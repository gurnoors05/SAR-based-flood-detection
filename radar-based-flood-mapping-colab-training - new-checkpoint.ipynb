```python
####################################################
#################### USER INPUT ####################
####################################################

# polarisations to be processed
Polarisation = 'VH'      # ["VH", "VV", "both"]

DownloadImage = True

# show intermediate results
PlotResults = True


#####################################################
###################### IMPORTS ######################
#####################################################

import sys
import os
import time
import json
import glob
import shutil
import functools

# SNAP
import esa_snappy as snappy
import jpy
from esa_snappy import ProductIO

# Geospatial
import geopandas
from osgeo import ogr, gdal, osr

# Sentinel API
from sentinelsat.sentinel import (
    SentinelAPI,
    read_geojson,
    geojson_to_wkt
)

# Widgets
from ipyfilechooser import FileChooser
import ipywidgets

# Visualization
import matplotlib.pyplot as plt
import numpy as np
import folium
from folium import plugins
from folium.plugins import MiniMap, Draw, Search

# Image processing
import skimage.filters

# Utilities
from datetime import date
from IPython.display import display
from zipfile import ZipFile


####################################################
############### FUNCTION DEFINITIONS ###############
####################################################

# Function looks for AOI file
def getAOI(path):
    try:
        file = readJSONFromAOI(path)
        print(f'AOI file found: {file}')
    except:
        raise FileNotFoundError(
            f'No AOI file found inside: {path}\n'
            'Please place a .geojson/.json/.shp/.kml/.kmz file there.'
        )

    return file


# Function looks for AOI file, converts to GeoJSON if needed
def readJSONFromAOI(path):

    # GeoJSON
    if len(glob.glob(f'{path}/*.geojson')) == 1:
        file = glob.glob(f'{path}/*.geojson')[0]

    elif len(glob.glob(f'{path}/*.json')) == 1:
        file = glob.glob(f'{path}/*.json')[0]

    # SHP → GeoJSON
    elif len(glob.glob(f'{path}/*.shp')) == 1:

        file_name = os.path.splitext(
            glob.glob(f'{path}/*.shp')[0]
        )[0].split('\\')[-1]

        shp_file = geopandas.read_file(
            glob.glob(f'{path}/*.shp')[0]
        )

        shp_file.to_file(
            f'{path}/{file_name}.json',
            driver='GeoJSON'
        )

        file = glob.glob(f'{path}/*.json')[0]

    # KML → GeoJSON
    elif len(glob.glob(f'{path}/*.kml')) == 1:

        file_name = os.path.splitext(
            glob.glob(f'{path}/*.kml')[0]
        )[0].split('\\')[-1]

        kml_file = gdal.OpenEx(
            glob.glob(f'{path}/*.kml')[0]
        )

        ds = gdal.VectorTranslate(
            f'{path}/{file_name}.json',
            kml_file,
            format='GeoJSON'
        )

        del ds

        file = glob.glob(f'{path}/*.json')[0]

    # KMZ → GeoJSON
    elif len(glob.glob(f'{path}/*.kmz')) == 1:

        with ZipFile(glob.glob(f'{path}/*.kmz')[0], 'r') as kmz:

            folder = os.path.splitext(
                glob.glob(f'{path}/*.kmz')[0]
            )[0]

            kmz.extractall(folder)

        if len(glob.glob(f'{folder}/*.kml')) == 1:

            kml_file = gdal.OpenEx(
                glob.glob(f'{folder}/*.kml')[0]
            )

            ds = gdal.VectorTranslate(
                f'{path}/{folder.split("/")[-1]}.json',
                kml_file,
                format='GeoJSON'
            )

            del ds

            file = glob.glob(f'{path}/*.json')[0]

            shutil.rmtree(folder)

    else:
        raise FileNotFoundError

    return file


# Plot raster band
def plotBand(band, threshold=None, binary=False):

    vmin, vmax = 0, 1

    w = band.getRasterWidth()
    h = band.getRasterHeight()

    band_data = np.zeros(w * h, np.float32)

    band.readPixels(0, 0, w, h, band_data)

    band_data.shape = h, w

    if binary:
        cmap = plt.get_cmap('binary')
    else:
        vmin = np.percentile(band_data, 2.5)
        vmax = np.percentile(band_data, 97.5)
        cmap = plt.get_cmap('gray')

    if threshold is None:

        fig, ax = plt.subplots(1, 1, figsize=(12, 12))

        ax.imshow(
            band_data,
            cmap=cmap,
            vmin=vmin,
            vmax=vmax
        )

        ax.set_title(band.getName())

    else:

        fig, (ax1, ax2) = plt.subplots(
            1, 2,
            figsize=(16, 6)
        )

        ax1.imshow(
            band_data,
            cmap=cmap,
            vmin=vmin,
            vmax=vmax
        )

        ax1.set_title(band.getName())

        band_data.shape = h * w

        ax2.hist(
            np.asarray(
                band_data[band_data != 0],
                dtype='float'
            ),
            bins=2048
        )

        ax2.axvline(x=threshold, color='r')

        ax2.set_title(
            f'Histogram: {band.getName()}'
        )

    for ax in fig.get_axes():
        ax.label_outer()


####################################################
####################### CODE #######################
####################################################

# LOCAL WORKING DIRECTORY
directory = r'E:\SAR Satellite Imagery\radar-based-flood-mapping'

if not os.path.isdir(directory):
    os.mkdir(directory)

# Folium basemaps
basemaps = {

    'Google Satellite Hybrid': folium.TileLayer(
        tiles='https://mt1.google.com/vt/lyrs=y&x={x}&y={y}&z={z}',
        attr='Google',
        name='Google Satellite',
        overlay=True,
        control=True,
        show=False
    ),

    'Google Satellite': folium.TileLayer(
        tiles='https://mt1.google.com/vt/lyrs=s&x={x}&y={y}&z={z}',
        attr='Google',
        name='Google Satellite',
        overlay=True,
        control=True,
        show=False
    )
}

print("Initialization successful!")
```

## Processing

<img src="https://github.com/vhertel/radar-based-flood-mapping/blob/main/resources/charts/chart3.png?raw=1" width="1000"/>

If more than one Sentinel-1 image exists in the *'input'* subfolder, the user can select which one is to be used for the processing. The subset is generated according to the AOI file in the *'AOI'* subfolder. If no AOI file is provided, an interactive map allows to draw and download the area of interest by clicking the *'Export'*-button inside the map or to directly upload a locally stored AOI file. Subsequently, the following processing steps are performed:

1. ***Apply Orbit File***: The orbit file provides accurate satellite position and velocity information. Based on this information, the orbit state vectors in the abstract metadata of the product are updated. The precise orbit files are available days-to-weeks after the generation of the product. Since this is an optional processing step, the tool will continue the workflow in case the orbit file is not yet available to allow rapid mapping applications.

```python
####################################################
################ SAFE STARTUP ######################
####################################################

import os

# IMPORTANT:
# Must be BEFORE importing esa_snappy
os.environ['_JAVA_OPTIONS'] = '-Xmx6g'

####################################################
#################### IMPORTS #######################
####################################################

import glob
import json
import time

import numpy as np
import folium
import esa_snappy as snappy
import jpy

from shapely.geometry import shape
from shapely.wkt import dumps as wkt_dumps
from IPython.display import display

####################################################
################ USER SETTINGS #####################
####################################################

directory = r'E:\SAR Satellite Imagery\radar-based-flood-mapping'

Polarisation = 'VH'

####################################################
################ SOURCE BANDS ######################
####################################################

if Polarisation == 'VH':
    sourceBands = 'Intensity_VH'

elif Polarisation == 'VV':
    sourceBands = 'Intensity_VV'

else:
    sourceBands = (
        'Intensity_VH,Intensity_VV'
    )

####################################################
################ HELPER FUNCTION ###################
####################################################

def geojson_to_wkt_shapely(data_json):

    geom = shape(
        data_json["features"][0]["geometry"]
    )

    return wkt_dumps(
        geom,
        rounding_precision=6
    )

####################################################
################ FIND AOI FILE #####################
####################################################

aoi_dir = os.path.join(directory, 'AOI')

geojson_files = glob.glob(
    os.path.join(aoi_dir, '*.geojson')
)

if len(geojson_files) == 0:
    raise FileNotFoundError(
        f'No GeoJSON found inside:\n{aoi_dir}'
    )

aoi_file = geojson_files[0]

print(f'AOI found:\n{aoi_file}')

####################################################
################ FIND SAFE FILE ####################
####################################################

input_dir = os.path.join(directory, 'input')

safe_files = glob.glob(
    os.path.join(input_dir, '*.SAFE')
)

if len(safe_files) == 0:
    raise FileNotFoundError(
        f'No .SAFE folders found inside:\n{input_dir}'
    )

safe_path = safe_files[0]

print(f'\nSAFE selected:\n{safe_path}')

####################################################
################ READ SAFE PRODUCT #################
####################################################

print('\nReading SAFE product...\n')

S1_source = snappy.ProductIO.readProduct(
    safe_path
)

if S1_source is None:
    raise RuntimeError(
        'SNAP could not read SAFE product.'
    )

print('SAFE product loaded')

####################################################
################ LOAD AOI ##########################
####################################################

with open(aoi_file, 'r') as f:

    data_json = json.load(f)

footprint = geojson_to_wkt_shapely(
    data_json
)

####################################################
################ APPLY SUBSET ######################
####################################################

print('\nApplying AOI subset...\n')

start_time = time.time()

parameters = snappy.HashMap()

parameters.put(
    'copyMetadata',
    True
)

parameters.put(
    'geoRegion',
    snappy.WKTReader().read(footprint)
)

parameters.put(
    'sourceBands',
    sourceBands
)

S1_crop = snappy.GPF.createProduct(
    'Subset',
    parameters,
    S1_source
)

print(
    'Subset complete '
    f'--- {time.time() - start_time:.2f} seconds'
)

####################################################
############### APPLY ORBIT FILE ###################
####################################################

print('\n1. Apply Orbit File...\n')

start_time = time.time()

parameters = snappy.HashMap()

parameters.put(
    'continueOnFail',
    True
)

S1_Orb = snappy.GPF.createProduct(
    'Apply-Orbit-File',
    parameters,
    S1_crop
)

print(
    'Orbit correction complete '
    f'--- {time.time() - start_time:.2f} seconds'
)

####################################################
################ VERIFY OUTPUT #####################
####################################################

print('\nBands after orbit correction:\n')

print(
    list(S1_Orb.getBandNames())
)

print('\nSUCCESS')
print('Orbit corrected product is stored IN MEMORY')
print('DO NOT save using ProductIO.writeProduct')
```

2. ***Thermal Noise Removal***: Thermal noise correction is applied to Sentinel-1 Level-1 GRD products which have not already been corrected.

```python
####################################################
# 2. THERMAL NOISE REMOVAL
####################################################

print('\n2. Thermal Noise Removal...\n')

start_time = time.time()

parameters = snappy.HashMap()

parameters.put(
    'removeThermalNoise',
    True
)

# Apply thermal noise removal
S1_Thm = snappy.GPF.createProduct(
    'ThermalNoiseRemoval',
    parameters,
    S1_Orb
)

print(
    'Thermal noise removal complete '
    f'--- {time.time() - start_time:.2f} seconds'
)

####################################################
################ VERIFY OUTPUT #####################
####################################################

print('\nBands after thermal removal:\n')

print(
    list(S1_Thm.getBandNames())
)

print('\nSUCCESS')
print('Thermal corrected product stored IN MEMORY')
```

3. ***Radiometric Calibration***: The objective of SAR calibration is to provide imagery in which the pixel values can be directly related to the radar backscatter of the scene. Though uncalibrated SAR imagery is sufficient for qualitative use, calibrated SAR images are essential to the quantitative use of SAR data.

```python
####################################################
# 3. RADIOMETRIC CALIBRATION
####################################################

print('\n3. Radiometric Calibration...\n')

start_time = time.time()

parameters = snappy.HashMap()

# Output sigma band
parameters.put(
    'outputSigmaBand',
    True
)

# Input band
parameters.put(
    'sourceBands',
    'Intensity_VH'
)

# Polarisation
parameters.put(
    'selectedPolarisations',
    'VH'
)

# Keep linear scale
parameters.put(
    'outputImageScaleInDb',
    False
)

####################################################
# APPLY CALIBRATION
####################################################

S1_Cal = snappy.GPF.createProduct(
    'Calibration',
    parameters,
    S1_Thm
)

print(
    'Calibration complete '
    f'--- {time.time() - start_time:.2f} seconds'
)

####################################################
################ VERIFY OUTPUT #####################
####################################################

print('\nBands after calibration:\n')

print(
    list(S1_Cal.getBandNames())
)

print('\nSUCCESS')
print('Radiometrically calibrated product stored IN MEMORY')
```

4. ***Speckle Filtering***: SAR images have inherent texturing called speckles which degrade the quality of the image and make interpretation of features more difficult. Speckles are caused by random constructive and destructive interference of the de-phased but coherent return waves scattered by the elementary scatter within each resolution cell. Speckle noise reduction can be applied either by spatial filtering or multilook processing. A Lee filter with an X, Y size of 5, 5 is used in this step.


```python
####################################################
# 4. SPECKLE FILTERING
####################################################

print('\n4. Speckle Filtering...\n')

start_time = time.time()

parameters = snappy.HashMap()

# Lee filter
parameters.put(
    'filter',
    'Lee'
)

# Smaller window for lower RAM
parameters.put(
    'filterSizeX',
    '3'
)

parameters.put(
    'filterSizeY',
    '3'
)

####################################################
# APPLY FILTER
####################################################

S1_Spk = snappy.GPF.createProduct(
    'Speckle-Filter',
    parameters,
    S1_Cal
)

print(
    'Speckle filtering complete '
    f'--- {time.time() - start_time:.2f} seconds'
)

####################################################
# CONVERT TO dB
####################################################

print('\nConverting to dB...\n')

S1_Spk_db = snappy.GPF.createProduct(
    'LinearToFromdB',
    snappy.HashMap(),
    S1_Spk
)

####################################################
################ VERIFY OUTPUT #####################
####################################################

print('Bands after speckle filtering:\n')

print(
    list(S1_Spk_db.getBandNames())
)

print('\nSUCCESS')
print('Speckle filtered product stored IN MEMORY')
```

5. ***Terrain Correction***: Due to topographical variations of a scene and the tilt of the satellite sensor, distances can be distorted in the SAR images. Data which is not directly directed towards the sensor’s Nadir location will have some distortion. Therefore, terrain corrections are intended to compensate for these distortions to allow a realistic geometric representation in the image.

```python
####################################################
# 5. TERRAIN CORRECTION
####################################################

print('\n5. Terrain Correction...\n')

start_time = time.time()

parameters = snappy.HashMap()

####################################################
# DEM SETTINGS
####################################################

parameters.put(
    'demName',
    'SRTM 1Sec HGT'
)

####################################################
# RESAMPLING
####################################################

parameters.put(
    'demResamplingMethod',
    'BILINEAR_INTERPOLATION'
)

parameters.put(
    'imgResamplingMethod',
    'NEAREST_NEIGHBOUR'
)

####################################################
# PIXEL SPACING
####################################################

parameters.put(
    'pixelSpacingInMeter',
    10.0
)

####################################################
# SAVE SOURCE BAND
####################################################

parameters.put(
    'saveSelectedSourceBand',
    True
)

####################################################
# SEA HANDLING
####################################################

parameters.put(
    'nodataValueAtSea',
    False
)

####################################################
# APPLY TERRAIN CORRECTION
####################################################

S1_TC = snappy.GPF.createProduct(
    'Terrain-Correction',
    parameters,
    S1_Spk_db
)

print(
    'Terrain correction complete '
    f'--- {time.time() - start_time:.2f} seconds'
)

####################################################
################ VERIFY OUTPUT #####################
####################################################

print('\nBands after terrain correction:\n')

print(
    list(S1_TC.getBandNames())
)

print('\nSUCCESS')
print('Terrain corrected product stored IN MEMORY')
```

6. ***Binarization***: In order to obtain a binary flood mask, the histogram is analyzed to separate water from non-water pixels. Due to the side-looking geometry of SAR sensors and the comparably smooth surface of water, only a very small proportion of backscatter is reflected back to the sensor leading to comparably low pixel values in the histogram. The threshold used for separation is automatically calculated using <a href="https://scikit-image.org/">scikit-image</a> 

7. ***Speckle Filtering***: A Median filter with an X, Y size of 7, 7 is used in this step.

```python
####################################################
# 6. FLOOD MASK + SAFE GEOTIFF EXPORT
####################################################

import os
import time
import numpy as np
import rasterio
from rasterio.transform import from_origin
import jpy

####################################################
# ADD LAND COVER
####################################################

print('\n6. Flood Mask Creation...\n')

start_time = time.time()

parameters = snappy.HashMap()

parameters.put(
    'landCoverNames',
    'GlobCover'
)

GlobCover = snappy.GPF.createProduct(
    'AddLandCover',
    parameters,
    S1_TC
)

print('GlobCover added')

print('\nBands available:\n')
print(list(GlobCover.getBandNames()))

####################################################
# FIXED THRESHOLD
####################################################

threshold = -17

print(f'\nUsing threshold: {threshold} dB')

####################################################
# BAND MATH
####################################################

BandDescriptor = jpy.get_type(
    'org.esa.snap.core.gpf.common.BandMathsOp$BandDescriptor'
)

targetBands = jpy.array(
    'org.esa.snap.core.gpf.common.BandMathsOp$BandDescriptor',
    1
)

targetBand = BandDescriptor()

targetBand.name = 'FloodMask'

targetBand.type = 'float32'

targetBand.expression = (
    f'if (Sigma0_VH_db < {threshold} '
    f'&& land_cover_GlobCover != 210) '
    f'then 1 else 0'
)

targetBands[0] = targetBand

parameters = snappy.HashMap()

parameters.put(
    'targetBands',
    targetBands
)

####################################################
# CREATE FLOOD MASK
####################################################

S1_floodMask = snappy.GPF.createProduct(
    'BandMaths',
    parameters,
    GlobCover
)

print(
    '\nFlood mask created '
    f'--- {time.time() - start_time:.2f} seconds'
)

####################################################
# READ PIXELS SAFELY
####################################################

band = S1_floodMask.getBand('FloodMask')

width  = band.getRasterWidth()
height = band.getRasterHeight()

print(f'\nRaster size: {width} x {height}')

####################################################
# GET GEOREFERENCING
####################################################

geoCoding = S1_floodMask.getSceneGeoCoding()

pos1 = geoCoding.getGeoPos(
    snappy.PixelPos(0, 0),
    None
)

pos2 = geoCoding.getGeoPos(
    snappy.PixelPos(1, 1),
    None
)

pixel_width  = pos2.lon - pos1.lon
pixel_height = pos1.lat - pos2.lat

transform = from_origin(
    pos1.lon,
    pos1.lat,
    pixel_width,
    pixel_height
)

####################################################
# OUTPUT FILE
####################################################

output_tif = os.path.join(
    directory,
    'output',
    'flood_mask_final.tif'
)

####################################################
# STRIP-WISE EXPORT
####################################################

print('\nWriting GeoTIFF safely...\n')

strip_size = 512

with rasterio.open(
    output_tif,
    'w',
    driver='GTiff',
    height=height,
    width=width,
    count=1,
    dtype=np.float32,
    crs='EPSG:4326',
    transform=transform,
) as dst:

    for y in range(0, height, strip_size):

        h = min(strip_size, height - y)

        data = np.zeros(
            width * h,
            np.float32
        )

        band.readPixels(
            0,
            y,
            width,
            h,
            data
        )

        data = data.reshape(h, width)

        dst.write(data, 1, window=((y, y+h), (0, width)))

        print(
            f'Written rows {y} → {y+h}',
            flush=True
        )

####################################################
# DONE
####################################################

print('\nSUCCESS')
print(f'Flood mask saved at:\n{output_tif}')
```

The processed flood mask is exported as GeoTIFF, SHP, KML, and GeoJSON and stored in the *'output'* subfolder. An interactive map shows the flood mask.

```python
import rasterio
import numpy as np

from scipy.ndimage import (
    binary_opening,
    binary_closing
)

####################################################
# PATHS
####################################################

input_tif = (
    r'E:\SAR Satellite Imagery\radar-based-flood-mapping'
    r'\output\flood_mask_final.tif'
)

output_tif = (
    r'E:\SAR Satellite Imagery\radar-based-flood-mapping'
    r'\output\flood_mask_cleaned.tif'
)

####################################################
# PROCESS STRIP BY STRIP
####################################################

strip_size = 512

with rasterio.open(input_tif) as src:

    profile = src.profile

    profile.update(dtype=rasterio.uint8)

    with rasterio.open(
        output_tif,
        'w',
        **profile
    ) as dst:

        for y in range(0, src.height, strip_size):

            h = min(
                strip_size,
                src.height - y
            )

            ####################################################
            # READ SMALL WINDOW
            ####################################################

            window = (
                (y, y + h),
                (0, src.width)
            )

            data = src.read(
                1,
                window=window
            )

            ####################################################
            # CLEAN
            ####################################################

            binary = data > 0

            clean = binary_opening(
                binary,
                structure=np.ones((3,3))
            )

            clean = binary_closing(
                clean,
                structure=np.ones((5,5))
            )

            clean = clean.astype(np.uint8)

            ####################################################
            # WRITE STRIP
            ####################################################

            dst.write(
                clean,
                1,
                window=window
            )

            print(
                f'Processed rows {y} → {y+h}',
                flush=True
            )

####################################################
# DONE
####################################################

print('\nSUCCESS')
print(f'\nSaved:\n{output_tif}')
```

```python
####################################################
# RERUN WITH STRICTER THRESHOLD + STRONGER EDGE MASK
####################################################
import os
import numpy as np
import rasterio
import rasterio.windows
from rasterio.enums import Resampling
from scipy.ndimage import median_filter
import matplotlib.pyplot as plt

directory  = r'E:\SAR Satellite Imagery\radar-based-flood-mapping'
final_path = os.path.join(directory, 'output', 'flood_mask_final.tif')
clean_path = os.path.join(directory, 'output', 'flood_mask_cleaned.tif')

with rasterio.open(final_path) as src:
    profile = src.profile.copy()
    width   = src.width
    height  = src.height

profile.update(dtype='float32', nodata=0)

####################################################
# SETTINGS — tune these
####################################################
edge_buffer = 200   # was 80 — larger to remove top-right block
strip_h     = 200
overlap     = 10

print('Cleaning with stricter settings...\n')

with rasterio.open(final_path) as src:
    with rasterio.open(clean_path, 'w', **profile) as dst:
        for row_start in range(0, height, strip_h):
            row_end    = min(row_start + strip_h, height)
            read_start = max(0, row_start - overlap)
            read_end   = min(height, row_end + overlap)

            chunk = src.read(
                1,
                window=rasterio.windows.Window(
                    0, read_start, width, read_end - read_start
                )
            ).astype(np.float32)

            # ── Stronger edge removal ─────────────
            if row_start < edge_buffer:
                chunk[:edge_buffer - row_start, :] = 0
            if row_end > height - edge_buffer:
                chunk[max(0, height - edge_buffer - read_start):, :] = 0
            chunk[:, :edge_buffer]  = 0
            chunk[:, -edge_buffer:] = 0

            # ── Also remove top-right corner block ─
            # Top 30% rows + right 40% cols = corner mask
            top_zone  = int(height * 0.30)
            right_zone = int(width * 0.60)
            if row_end < top_zone:
                chunk[:, right_zone:] = 0
            elif row_start < top_zone:
                partial = top_zone - row_start
                chunk[:partial, right_zone:] = 0

            # ── Median filter ─────────────────────
            filtered = median_filter(chunk, size=7)  # stronger: 7 not 5
            filtered = (filtered > 0.5).astype(np.float32)

            # Crop overlap
            crop_top = row_start - read_start
            crop_bot = crop_top + (row_end - row_start)
            filtered = filtered[crop_top:crop_bot, :]

            dst.write(
                filtered, 1,
                window=rasterio.windows.Window(
                    0, row_start, width, row_end - row_start
                )
            )

            pct      = row_end / height * 100
            flood_px = int(np.sum(filtered > 0))
            print(
                f'  {pct:5.1f}%  rows {row_start}–{row_end}'
                f'  flood px: {flood_px:,}',
                flush=True
            )
            del chunk, filtered

size_mb = os.path.getsize(clean_path) / (1024 * 1024)
print(f'\n✅ Saved: {clean_path}  ({size_mb:.1f} MB)')

####################################################
# PREVIEW
####################################################
with rasterio.open(clean_path) as src:
    preview = src.read(
        1,
        out_shape=(height // 8, width // 8),
        resampling=Resampling.nearest
    ).astype(np.float32)

flood_px  = int(np.sum(preview > 0))
flood_km2 = flood_px * 64 * 100 / 1e6
print(f'\nFlood pixels : {flood_px:,}')
print(f'Approx area  : {flood_km2:.1f} km²')

fig, ax = plt.subplots(figsize=(12, 10))
ax.imshow(preview, cmap='Blues', vmin=0, vmax=1)
ax.set_title(
    f'Flood Mask Cleaned — Kerala August 2018\n'
    f'Approx flooded area: {flood_km2:.0f} km²'
)
ax.axis('off')
plt.tight_layout()
plt.show()
```
